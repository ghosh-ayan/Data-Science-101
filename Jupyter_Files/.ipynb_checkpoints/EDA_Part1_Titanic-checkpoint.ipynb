****
***Edit #001***: What I was trying to accomplish above can be done in a much better way using seaborn countplot. No need to find out value_counts and then plot on it, create titles etc.
Use sns.countplot() and just pass the pandas feature name as the x-parameter. Since this is a countplot, y axis will always be the count. Therefore, for categorical features if we want to plot a barchart showing counts corresponding to each values of that feature, we use countplot. This is a special kind of a barplot but no statistical in nature like a barplot
****

```python
# Create the figure and the axes
plt.rcParams.update({'axes.labelcolor': '#4934eb'})
fig1, axs1 = plt.subplots(4, 2, figsize=(10,10))

# Adding the plots to each of the axes
sns.countplot(x="sex", palette="Greens", data=titanic, ax = axs1[0, 0])
sns.countplot(x="embarked", palette="BuGn", data=titanic, ax = axs1[0, 1])
sns.countplot(x="class", palette="OrRd", data=titanic, ax = axs1[1, 0])
sns.countplot(x="who", palette="PuOr", data=titanic, ax = axs1[1, 1])
sns.countplot(x="deck", palette="ch:.25", data=titanic, ax = axs1[2, 0])
sns.countplot(x="embark_town", palette="PuRd", data=titanic, ax = axs1[2, 1])
sns.countplot(x="alive", palette="RdYlGn", data=titanic, ax = axs1[3, 0])
sns.countplot(x="survived", palette="Reds", data=titanic, ax = axs1[3, 1])
#plt.gcf().get_axes()[0].set_title('test')
# Add a title to the figure
fig1.suptitle("Categorical Value Counts",size=15,weight=700, color='#eb8f34')
# Show the figure
#plt.tight_layout() --> replaced by constrained layout as a better option to prevent overlapping of axes and figure titles
plt.show()
```

```python
plt.rcParams['figure.constrained_layout.use'] = True
fig, axs = plt.subplots(2, 4, figsize=(10,5))
sns.barplot(x='sex', y='survived', data=titanic, ax=axs[0, 0])
sns.barplot(x='who', y='survived', data=titanic, ax=axs[0, 1])
sns.barplot(x='alone', y='survived', data=titanic, ax=axs[0, 2])
sns.barplot(x='pclass', y='survived', data=titanic, ax=axs[0, 3])
sns.barplot(x='deck', y='survived', data=titanic, ax=axs[1, 0])
sns.barplot(x='embark_town', y='survived', data=titanic, ax=axs[1, 1])
plt.sca(axs[1, 1])
plt.xticks(rotation=90)
sns.barplot(x='adult_male', y='survived', data=titanic, ax=axs[1, 2])
axs[1,3].set_visible(False)
```

What we know from the above plots are:-
- Females survived way lot more than males
- The % of children survived are way more than % of men survived. Therefore, adult males were more likely to die than male children, as apparent from the adult_male plot
- People who were with someone on the ship were more likely to survive.
- The likelihood of survival based on passenger class in the following order - first class > second class > third class
- Deck does not seem to accurately determine the likelihood of survival on a cursory glance. But we need to take into account the physical construction of the ship, and which parts may have had more and rapid damage from the iceberg collision (this is extraneous knowledge, of course, not available in the dataset, but using extraneous investigation and incorporating the business/domain knowledge will always improve predictions)


Below value_counts show that barplot() above does not exactly work in terms of counts. In 'who' chart above, the bar corresponding to child
is higher than the bar for men. It does not mean that the count of men who survived is lower than the count of children who survived. But it means that the proportion of surviving children is higher. Therefore the chart gives us exactly what we need - relative likelihood of survival among each of the values under each feature. Barplot is a statistical plot (See ***Edit 001*** above regarding ***sns.countplot*** )

```python
titanic.loc[titanic.alone == True, 'survived'].value_counts(normalize=True)
```

```python
titanic.loc[titanic.alone == False, 'survived'].value_counts(normalize=True)
```

As also shown in the barplot for 'alone', the above normalized counts also suggest that - If you were not alone, your chances of survival are 50-50. But if you were alone, your chances are about 70-30 in favour of death.

## Statistical analysis

```python
titanic.describe()
```

Age and fare are the only real continuous numeric variables on which the statistical measures make sense.
- Average age of the people aboard the ship was 29 years. But the dispersion among age is high (14.5). There were babies(min age 0.42) on the ship and the maximum age was 80.
- There were people who had 0 fare (need to check is this really zero or do they signify missing value). Max was 512 pounds. Average fare is 32 pounds. standard deviation among fares is about 50 pounds. Also, 75% of the population had fares lower than 31 pounds (in fact 50% paid less than 15 pounds), but the fares really shot up in the topmost 25% which were probably some of the best in the first class.
