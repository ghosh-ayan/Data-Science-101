## Installing Jupyterthemes and Nbextensions
<p>
1. To install jupyterthemes, On the anaconda command prompt -

<code> conda install -c conda-forge jupyterthemes </code>
2. Re-launch JupyterNb. On the Jupyter Notebook on any cell we can type below to get all the available themes-

<code>! jt -l</code>
Available Themes: 
   chesterish
   grade3
   gruvboxd
   gruvboxl
   monokai
   oceans16
   onedork
   solarizedd
   solarizedl
3. Use monokai/chesterish theme with below specifications, second one is bitstream font and cell width reduced to 85%.

<code>! jt -t monokai -f fira -fs 13 -nf ptsans -nfs 11 -N -kl -cursw 5 -cursc r -cellw 95% -T</code>
 
<code>! jt -t chesterish -f bitstream -fs 12 -nf bitstream -nfs 11 -N -kl -cursw 5 -cursc r -cellw 85% -T </code>

4. To install notebook extensions, On the anaconda command prompt -

<code>conda install -c conda-forge jupyter_contrib_nbextensions</code>

5. Re-launch JupyterNb. Go to Nbextensions tab and enable the following extensions -

<code>Hinterland, Autopep8, Table Beautifier, Scratchpad</code>

6a . So that the plots are not getting messed up due to the theme,use below as an import statement
in the notebook or 6b-

<code>from jupyterthemes import jtplot
jtplot.style(theme='monokai', context='notebook', ticks=True, grid=False)</code>

6b. Make a file named 00_startup.py in ~/.ipython/profile_default/startup and stick the following snippet
into it, and restart Jupyter :-

<code>import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from jupyterthemes import jtplot
jtplot.style(theme='monokai', context='notebook', ticks=True, grid=False)</code>

Note that you can use one theme for the notebook and another for jtplot. I am using grade3 for jtplot to
have the plots displayed as close to the default theme as possible. Just a matter of preference.
Read and try with others. Make sure you do not unnecessarily slow down your system.
</p>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from jupyterthemes import jtplot
jtplot.style(theme='grade3', context='notebook', ticks=True, grid=False)
```

```python
a = sns.load_dataset("flights")
```

```python
sns.relplot(x='passengers', y='month', data=a)
```
