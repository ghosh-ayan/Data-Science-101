```python
'''
You can use Slicing or indexing on the columns attribute to get a list of column labels. This list can be used to drop columns 
collectively without writing names of each and every column on pd.DataFrame.drop method.
I am retrieving the column names as a series just so that I can easily know the index number of the columns I want to
delete using : slicing. It is difficult to mention all the comma separated column label names, or even one or two label
names with slice operator (as the label names are large as you can see below)
Also the pd.set_option is being used so that I can see all the 83 rows on the resulting series(which are nothing but
83 column labels in the dataframe). 
This option can be reset later with pd.reset_option('display.max_rows').
'''
pd.set_option('display.max_rows',None)
cols = pd.Series(dft.columns.values)
cols
```

```python
dft.drop(dft.columns[10:56], axis=1, inplace=True)
cols = pd.Series(dft.columns.values)
cols
```

```python
dft.drop(dft.columns[15:35], axis=1, inplace=True)
cols = pd.Series(dft.columns.values)
cols
```

```python
'''
This is an incredible piece of code that gives you a Series of the number of missing values indexed by each and every column
of the dataframe. df.isna() returns a dataframe of booleans of same shape as the original dataframe, having False where it 
is not NaN and True where it is NaN
Based on below one can take a decision later on how to, or whether or not to, handle these missing values appropriately.
'''
dft.isna().sum()
```
