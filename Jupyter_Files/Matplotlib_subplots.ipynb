```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic', data_home='../../Data')
```

*** <code> fig, axs = plt.subplots(nrows, ncols, etc) </code> ***  
No need to separately add variables for figure and individual subplots  
The above statement gives us all the figure and axes variables we need. We can then either call axs[].plot() or
use the dataframe and series plot() method and specify the axes through the 'ax' parameter

```python
fig1, axs1 = plt.subplots(8,figsize=(15,30))
titanic.sex.value_counts().plot(kind='barh', grid=True, ax=axs1[0],title='sex')
titanic.embarked.value_counts().plot(kind='barh', grid=True, ax=axs1[1],title='embarked')
titanic['class'].value_counts().plot(kind='barh', grid=True, ax=axs1[2],title='class')
titanic.who.value_counts().plot(kind='barh', grid=True, ax=axs1[3],title='who')
titanic.deck.value_counts().plot(kind='barh', grid=True, ax=axs1[4],title='deck')
titanic.embark_town.value_counts().plot(kind='barh', grid=True, ax=axs1[5],title='embark_town')
titanic.alive.value_counts().plot(kind='barh', grid=True, ax=axs1[6],title='alive')
titanic.survived.value_counts().plot(kind='barh', grid=True, ax=axs1[7],title='survived')
fig1.suptitle("Categorical Value Counts",size=15,weight=700)
plt.show()
```


 ***plt.subplot()***  
 <code> 
    fig1 = plt.figure(1) 
    ax1 = plt.subplot(121)
    ax1.plot(X,Y)
    ax2 = plt.subplot(122)
    ax2.plot(X,Y)
 </code>

Instead of fig.add_subplot(111), uses plt.subplot(111)   
Actually they are more or less one and the same.   
add_subplot() method is called on a figure, that allows us to specify which figure to add the subplot to.   
plt.subplot() function directly adds subplot to the current figure.

```python
fig2 = plt.figure(figsize=(15,30))
ax1 = plt.subplot(811)
titanic.sex.value_counts().plot(kind='barh', grid=True, ax=ax1,title='sex')
ax2 = plt.subplot(812)
titanic.embarked.value_counts().plot(kind='barh', grid=True, ax=ax2,title='embarked')
ax3 = plt.subplot(813)
titanic['class'].value_counts().plot(kind='barh', grid=True, ax=ax3,title='class')
ax4 = plt.subplot(814)
titanic.who.value_counts().plot(kind='barh', grid=True, ax=ax4,title='who')
ax5 = plt.subplot(815)
titanic.deck.value_counts().plot(kind='barh', grid=True, ax=ax5,title='deck')
ax6 = plt.subplot(816)
titanic.embark_town.value_counts().plot(kind='barh', grid=True, ax=ax6,title='embark_town')
ax7 = plt.subplot(817)
titanic.alive.value_counts().plot(kind='barh', grid=True, ax=ax7,title='alive')
ax8 = plt.subplot(818)
titanic.survived.value_counts().plot(kind='barh', grid=True, ax=ax8,title='survived')
fig2.suptitle("Categorical Value Counts",size=15,weight=700)
plt.show()
```
