```python
Idvars = df_copy.columns[:10]
Valuevars = df_copy.columns[10:] 

df_melt = pd.melt(df_copy,
                  id_vars = Idvars,
                  value_vars = Valuevars,
                  value_name = 'Hours'
                 )
df_melt.head()
```

```python
colsplit = df_melt['variable'].str.split(' ', 1)
#colsplit.str[0] gives the series of first words in the entire colsplit series, 
#colsplit[0] would have given the first element of colsplit series which is a list.
#When we need to do something to the texts inside all the elements in the series, we use str
#.str just lets us do the same thing on the whole series.
df_melt['Effort Type'] = colsplit.str[0]
df_melt['Task Name'] = colsplit.str[1]
df_melt.head()
```

```python
#df_melt['Task Name'] = df_melt['Task Name'].str.split(' ', -1)'
#df_melt['Task Name'].str.rsplit(' ', 1)
df_melt['Task Name']= df_melt['Task Name'].str.rstrip('Hours')
```

```python
df_melt.drop('variable', axis=1, inplace=True)
```

```python
df_melt.info()
```

```python
df_melt.rename(columns={'Planned Hours':'Total Planned Hours',
                        'Actual Hours':'Total Actual Hours',
                        'Remaining Hours':'Total Remaining Hours'},
               inplace=True)
df_melt['AGID'] = df_melt['Resource Name'].str[-8:-1]
```

```python
df_melt['Resource Name'] = df_melt['Resource Name'].str[:-9]
df_melt
```

## Accessing particular sections of the dataframe
- To access specific Columns, we can select by *df[colname]* or *df[[colname1, colname2]]*. This will give the series.
- To access specific rows, we can select by df.loc[indexname] or df.loc[[indexname1, indexname2]]

```python
# Selecting one column --> Gives series
df_play['AGID']
```

```python
# Selecting more than one column --> Gives dataframe
df_play[['AGID','Resource Name']]
```

```python
# Selecing one or more rows by slicer based indexing. Endpoints excluded. Always returns dataframe as it is a slice
# Note, a single row cannot be selected by simply putting df_play[0],df_play[1] etc, it has to be a slice in order to get
# rows
df_play[0:1]
```

.loc is a very versatile and powerful attribute. You can use it to get subsections of your dataframe. Typically loc is always followed by a way to identify rows. Then we can use next parameter to pass columns.

Filtering just means restricting the number of data rows (not restricting by column, all columns will be there)
Note that dataframe have label-based access. To filter the dataframe, we need to filter on basis
of row or column labels (something like df[0], df[1], df[0,1] - these don't mean anything and doesn't work, if you
want something like that you need to use .iloc to fiter rows by subscripting "row numbers")

To filter on basis of row labels (restrict the number of rows to specified rows), we use .loc

To filter on basis of column labels(restrict the number of rows to specified rows that meet the condition), we use
boolean indexing

Finally to restrict the output to have only certain columns, we can qualify the column list either before .loc/.iloc or after .loc/.iloc, it does not matter which one we choose

```python
# If I do not know the index names of the first 2 rows I cannot use .loc, instead I will use .iloc.
# In below example df_play.iloc[[1,2]] works same as df_play.loc[[563,1833]]

df_play.iloc[[1,2]]
```

```python
df_play.iloc[1:2]
```
